```python
def init_browser():
    # @NOTE: Replace the path with your actual path to the chromedriver
    executable_path = {"executable_path": "D:\chromedriver"}
    return Browser("chrome", **executable_path, headless=False)

    browser = init_browser()
    listings = {}

    url = "https://www.google.com/maps/dir/618+Ansel+Road,+Burlingame,+CA+94010-3830,+USA/18+Alder+St,+San+Francisco,+CA+94134/@37.6514147,-122.4498165,12z/data=!3m1!4b1!4m14!4m13!1m5!1m1!1s0x808f9df4fd8552f3:0xaa7e458145b843bf!2m2!1d-122.3532396!2d37.5804316!1m5!1m1!1s0x808f7ede26567dbd:0x2a81f5d3ff1fe426!2m2!1d-122.4051253!2d37.7191144!3e0"
    browser.visit(url)

    html = browser.html
    soup = BeautifulSoup(html, "html.parser")
```
